```python
from selenium import webdriver  # Import Selenium library
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Initialize the WebDriver
# WebDriver-i başlatırıq
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# List of URLs to scrape
# Scrape edəcəyimiz URL-lərin siyahısı
urls = [
    "https://irshad.az/telefon-ve-aksesuarlar/mobil-telefonlar/xiaomi?price_range=169%2C2550",
    "https://irshad.az/telefon-ve-aksesuarlar/mobil-telefonlar/samsung",
]

# List to store all product names
# Bütün məhsul adlarını saxlayacaq siyahı
all_product_names = []

# Loop through each URL
# Hər URL üzərində dövr edirik
for url in urls:
    driver.get(url)  # Open the URL
    # URL-ni açırıq
    
    driver.implicitly_wait(30)  # Wait for the page to load
    # Sayfanın tam yüklənməsini gözləyirik
    
    try:
        # Wait for the "Daha çoxuna bax" button to be clickable
        # "Daha çoxuna bax" düyməsinin kliklənə bilən olmasını gözləyirik
        more_button = WebDriverWait(driver, 40).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, '.btn.btn-orange'))
        )
        
        # Click the button
        # Düyməyə tıklayırıq
        more_button.click()
        
        # Wait for the page to load after clicking the button
        # Butona tıkladıktan sonra sayfanın yenidən yüklənməsini gözləyirik
        time.sleep(5)
        
        # Click the button again
        # İkinci dəfə düyməyə tıklayırıq
        more_button.click()
        
        # Wait for the content to load
        # İçəriyi yüklənməsini gözləyirik
        time.sleep(5)
        
        # Wait for new product elements to be present
        # Yeni məhsul elementlərinin görünməsini gözləyirik
        product_elements = WebDriverWait(driver, 30).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.product__name'))
        )
        
        # Extract product names
        # Məhsul adlarını çıxarırıq
        product_names = [element.text for element in product_elements]
        
        # Append all product names to the list
        # Bütün məhsul adlarını siyahıya əlavə edirik
        all_product_names.extend(product_names)
    
    finally:
        # Close the browser
        # Brauzeri bağlayırıq
        driver.quit()

# Create a DataFrame with the product names
# Məhsul adları ilə DataFrame yaradırıq
df = pd.DataFrame(all_product_names, columns=['Product Name'])

# Save the DataFrame to a CSV file
# DataFrame-i CSV faylına yazırıq
df.to_csv('product_names.csv', index=False, encoding='utf-8')

# Inform the user
# İstifadəçiyə məlumat veririk
print("Ürün isimleri 'product_names.csv' dosyasına kaydedildi.")
# English: Inform the user that product names have been saved to 'product_names.csv'
print("Product names have been saved to 'product_names.csv'.")

```
